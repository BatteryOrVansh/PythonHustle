## SECTION A (BASIC PLOTS)

● Line Plot:
1. Generate a time series dataset (e.g., temperature readings over time).
2. Create a line plot to visualize the trend of the data.
3. Label the axes and add a title.

```python
import matplotlib.pyplot as plt

temp=[22,23,25,20,18,19,24]
days=['mon','tues','wed','thurs','fri','sat','sun']
plt.xlabel('Days of the week')
plt.ylabel('Temperature variation')
plt.title('TEMPERATURE OVER TIME')
plt.plot(days,temp)
```

● Bar Graph:
1. Simulate data representing website traffic for different days of the week.
2. Create a bar graph to compare the daily traffic.
3. Customize the bar colors and add legends.

```python
days=['mon','tues','wed','thurs','fri','sat','sun']
traffic=[2000,3800,1800,1600,2300,2500,2100]
plt.figure(figsize=(8,6))
plt.bar(days,traffic,color='orange',edgecolor='black')
plt.legend(['Traffic'])
plt.xlabel('Days of the week')
plt.ylabel('Website Traffic')
plt.title("WEBSITE TRAFFIC VARIATION")
plt.show()
```

● Histogram:
1. Generate random data representing exam scores.
2. Create a histogram to visualize the distribution of scores.
3. Adjust the number of bins and add labels for the x and y axes.

```python
import numpy as np

data={'Roll no.':np.arange(1,62),'Marks':np.random.randint(10,size=61)}
df=pd.DataFrame(data)
plt.hist(df.Marks,bins=10,edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')

plt.grid(True)
plt.show()
```

● Scatter Plot:
1. Simulate data for weight and height of individuals.
2. Create a scatter plot to explore the potential correlation between them.
3. Color code data points based on a third variable (e.g., gender).

```python
hm=[6,5.11,5.4,6.2,5]
wm=[80,70,69,82,77]

hf=[5.2,5.3,5,5.8,5.10]
wf=[49,52,55,65,63]

plt.scatter(hm,wm,color='blue',label='Male')
plt.scatter(hf,wf,color='pink',label='Female')

plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')

plt.legend(loc='lower right')
plt.show()
```

● Pie Chart:
1. Generate data representing the market share of different companies.
2. Create a pie chart to visualize the proportional distribution.
3. Add slice labels and percentages for better understanding.

```python

data={'companies':['reliance','tata','apple','asus','airtel'],
      'market_share':[24, 26, 15, 20,15]}
df=pd.DataFrame(data)

plt.pie(df['market_share'],labels=df['companies'],explode=(0.02,0.02,0.02,0.02,0.02),autopct=('%1.1f%%'))
plt.title('shares of companies')
plt.show()
```

● Polar Plot:
1. Generate data for wind speed and direction.
2. Create a polar plot to visualize the wind speed distribution across different directions.


```python
import numpy as np
import matplotlib.pyplot as plt

wd=np.arange(30,330,30)
ws=np.random.randint(5,20,10)
theta=np.radians(wd)
plt.figure(figsize=(10,8))
plt.polar(theta,ws,linestyle="-",color="green",marker="*",markersize=10,markerfacecolor="red")
plt.title("Polar Plot")
plt.xlabel("Wind direction")
plt.ylabel("Wind speed")
plt.show()
```

● Area Plot:
1. Simulate data for stock prices over time.
2. Create an area plot to visualize price fluctuations.
3. Optionally, add a secondary y-axis for another variable (e.g., trading volume).

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

stock=[100,30,75,50,150,125,160,90]
years=[2000,2001,2002,2003,2004,2005,2006,2007]
trading_volume=[14,16,19,15,25,20,17,22]

ax.fill_between(years, stock, color= 'blue',alpha=0.5,label="stock prices")
ax.tick_params(axis='y', labelcolor='blue')

ax2=ax.twinx()
ax2.fill_between(years, trading_volume, color='red', alpha=0.5, label='trading volume')
ax2.tick_params(axis='y', labelcolor='red')

ax.set_xlabel("Years")
ax.set_ylabel("stock prices", color='blue')
ax2.set_ylabel('trading volume',color='red')
plt.title('STOCK PRICES AND TRADING VOLUME \n OVER TIME')


plt.show()
```

● Bubble Chart:
1. Generate data for population, area, and GDP of different countries.
2. Create a bubble chart to visualize the relationship between these three variables.
3. Use bubble size to represent a fourth variable (e.g., population density).

```python

import matplotlib.pyplot as plt



population = [10,45,33,150,7,300] # in millions
area = [0.2,0.5,1.1,3.2,1,2.5] # in km sq
gdp = [0.8,1,1.2,5,0.6,3.3 ] # in trilion usd

pop_d=[pop/a for pop, a in zip(population, area)]

size=[]
for i in population:
    a=i*10
    size.append(a)

plt.figure(figsize=(8,6))
plt.scatter(gdp,population, s=size, c=pop_d )

cbar=plt.colorbar()
cbar.set_label('Population Density (per sq km)')

plt.xlabel('GDP (in Trillion USD)')
plt.ylabel('POPULATION (in Millions)')
plt.title("COUNTRIES' DATA")
plt.show()
```
