```python
# type conversion

x = "100"
y = int(x)  
z = float(x)
s = str(y)

print("x", x, "type:", type(x))
print("y", y, "type:", type(y))
print("z", z, "type:", type(z))
print("s", s, "type:", type(s))
```

```python
# Strings

text = "Python Programming"

print(text.upper())
print(text.lower())
print(text.replace("Python", "Ajgar"))
print(text)
print(text.split())
print(type(text.split()))
print(len(text))

```
