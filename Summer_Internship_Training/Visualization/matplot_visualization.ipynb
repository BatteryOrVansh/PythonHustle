```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
categories = ['A', 'B', 'C', 'D']   
values = [24, 46, 13, 38]

# find maximum value
max_value = max(values)

# assign colors
colors = ["green" if value != max_value else "orange" for value in values]

plt.bar(categories, values, color=colors)

plt.title("Bar Chart: Compare Categories", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.xlabel ("Value", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.ylabel ("Categories", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.show()
```

```python
domain = ["DS", "AI", "ML", "BC", "IOT", "DSA"]
cutoff = [85, 90, 85, 75, 70, 80]

plt.plot(domain, cutoff, color="black", marker="*")

plt.title("Line Chart: Cutoff Marks", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.xlabel("Domain")
plt.ylabel("Cutoff Marks")

plt.show()
```

```python
#2. Line chart - show trends over time

# Create a sequence of 30 consecutive dates starting from 20226-01-01
days = pd.date_range(start="2026-01-01", periods=30)

# Generate 30 random numbers from a normal distribution (mean=0, std=1)
# Compute their cumulative sum and shift all values upward by 50
values = np.cumsum(np.random.normal(0,1,30)) + 50

# Plot a line chart
plt.plot(
    days, 
    values,
    marker="o",
)

plt.title("Line Chart: Random Values Over Time")
plt.xlabel("Date")
plt.ylabel("Values")


# rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=45)

plt.show()
```

```python
#3. Pie chart - show percentage

labels = ["Product A", "Product B", "Product C", "Product D"]
sizes = [38,25,20,20]

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Pie Chart: Percentage share")
plt.show()
```

```python
#4. Histogram - show data distribution

data = np.random.normal(loc=60, scale=10, size=1000)
plt.hist(data, bins=30, color="green", edgecolor="red")
plt.title("Histogram: Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
