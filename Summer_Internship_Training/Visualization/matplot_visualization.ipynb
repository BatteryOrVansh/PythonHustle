```python
import matplotlib.pyplot as plt
```

```python
categories = ['A', 'B', 'C', 'D']   
values = [24, 46, 13, 38]

# find maximum value
max_value = max(values)

# assign colors
colors = ["green" if value != max_value else "orange" for value in values]

plt.bar(categories, values, color=colors)

plt.title("Bar Chart: Compare Categories", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.xlabel ("Value", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.ylabel ("Categories", fontdict={
    "fontsize": 16,
    "fontweight": "bold",
    "color": "red",
    "family": "Arial"
    })

plt.show()
```
