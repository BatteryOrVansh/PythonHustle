```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# First check list of functionality of seaborn
print(dir(sns))

# of matplotlib
print(dir(plt))
```

```python
ft = sns.load_dataset('flights')
```

```python
ft
```

```python
# Bar Plot (Average passengers by month)
sns.barplot(data=ft, x="month", y="passengers",)

plt.xticks(rotation=20)
plt.title("Average Passengers by Month")
plt.show()
```

```python
# Box Plot (Average Passengers by Month)
sns.boxplot(data=ft, x="month", y="passengers")
```

```python
# Violin Plot: Month wise passenger

sns.violinplot(data=ft, x="month", y="passengers")
```

```python
# Scatter Plot
sns.scatterplot(data=ft, x="year", y="passengers")
```

```python
# Histogram: Passengers
sns.histplot(ft["passengers"], bins=10)
plt.title("Passenger Distribution")
plt.show()
```

```python

# KDE Plot: Passengers
sns.kdeplot(ft["passengers"], fill=True)
plt.title("Density Plot")
plt.show()
```

```python
# Heatmap (Most popular plot for flights dataset)

pivot = ft.pivot(index="month", columns="year", values="passengers")
sns.heatmap(pivot, cmap="YlGnBu")
plt.title("Flights Heatmap")
plt.show()
```

```python
# Pair plot since the dataset has only one numeric featuer (passengers)
# the pair plot is simple

sns.pairplot(ft)
plt.show()
```

```python

# Count Plot: count the number of records for each month

sns.countplot(data=ft, x="month")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 8))

# 1. Line Plot
plt.subplot(2,2,1)
sns.lineplot(data=ft, x="year", y="passengers")
plt.show()
```

```python
# Line Plot with Hue
sns.lineplot(data=ft, x="year", y="passengers", hue="month")
plt.title("Monthly Passenger Trend")
plt.show()
```
