```python
import numpy as np
#original array
samplearray = np.array([[3, 6, 9, 12],
                       [15, 18, 21, 24],
                       [27, 30, 33, 36], 
                       [39, 42, 45, 48]])

#using slicing to select odd rows and even columns
#the slice [1::2] selects every second element starting from index 1 (odd rows)
#the slice [0::2] selects every second element starting from index 0 (even columns)
newArray = samplearray[1::2, 0::2]
print("Original array:")
print(samplearray)
print("\nNew array with odd rows and even columns:")
print(newArray)
```
