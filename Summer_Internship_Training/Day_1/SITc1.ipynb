```python
num1 = 10
num2 = 5

# Addition
sum_result = num1 + num2
print(f"Addition: {num1} + {num2} = {sum_result}")

# Subtraction
diff_result = num1 - num2
print(f"Subtraction: {num1} - {num2} = {diff_result}")

# Multiplication
prod_result = num1 * num2
print(f"Multiplication: {num1} * {num2} = {prod_result}")

# Division
div_result = num1 / num2
print(f"Division: {num1} / {num2} = {div_result}")

# Integer Division
int_div_result = num1 // num2
print(f"Integer Division: {num1} // {num2} = {int_div_result}")

# Modulo
mod_result = num1 % num2
print(f"Modulo: {num1} % {num2} = {mod_result}")
```

```python
data = {
    "Name" : ["Aman", "Riya", "Rahul", "Sneha", "Arjun", "Priya"],
    "Age" : [21, 22, 20, 23, 21, 22],
    "Marks" : [90, 85, 78, 92, 88, 76],
    "Hours_Study" : [3, 2, 4, 5, 3, 2]
}

df = pd.DataFrame(data)

print("Full Dataset: \n\n", df, sep="")
```

```python
print("First 5 rows: \n\n", df.head(), sep="")

print("Last 5 rows: \n\n", df.tail(3), sep="")
```

```python
# dataset information
print("dataset information\n\n", df.info(), sep="")
```

```python
# statistical description
print("statistical description\n\n", df.describe(), sep="")
```

```python
print("mean/avg of marks")
print(df["Marks"].mean())

print("\nmax of marks")
print(df["Marks"].max())

print("\nmin of marks")
print(df["Marks"].min())
```

```python
# filter
print("students scoring more than 85:")
print(df[df["Marks"] > 85])
```

```python
# correlation bw study hrs and marks
print("\nCorrelation:")
print(df[["Hours_Study", "Marks"]].corr())
print("\n")

# plot graph
plt.scatter(df["Hours_Study"], df["Marks"])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
