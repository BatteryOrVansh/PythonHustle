```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = {
    "Height" : [185, 170, 180, 175, 160, 155, 165, 190, 200, 195, 180, 177],
    "Weight" : [80, 60, 75, 70, 55, 50, 65, 90, 100, 95, 67, 76]
}
df = pd.DataFrame(data)
df
```

```python
#Plot dataset
plt.figure(figsize=(7,6))

plt.scatter(df["Height"], df["Weight"], s=120)

for i in range(len(df)):
    plt.text(df["Height"][i]+0.2,
             df["Weight"][i]+0.2,
             str(i+1))

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Original Dataset")
plt.grid(True)

plt.show()
```

```python
# Initial Centroids
initial_centroids = np.array([[160, 50], [200, 100]])
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1)
```
