```python
import pandas as pd

# Dictionary
data = {
    "Actual": [
        "Disease","Disease","Disease","Disease","Disease",
        "Disease","Disease","Disease",
        "No Disease","No Disease","No Disease","No Disease",
        "No Disease","No Disease","No Disease","No Disease",
        "No Disease","No Disease","No Disease","No Disease"
    ],

    "Predicted": [
        "Disease","Disease","Disease","No Disease","Disease",
        "No Disease","Disease","Disease",
        "Disease","No Disease","No Disease","No Disease",
        "Disease","No Disease","No Disease","No Disease",
        "No Disease","Disease","No Disease","No Disease"
    ]
}

df = pd.DataFrame(data)

print(df)
df.shape
```

|                | Predicted Disease | Predicted Healthy |
| -------------- | ----------------- | ----------------- |
| Actual Disease | TP=6              | FN=2              |
| Actual Healthy | FP=3              | TN=9              |


```python
#Import Evaluation Metrics

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    classification_report,
    roc_auc_score
)
```

```python
#Convert Labels into Binary Values

# Disease = 1
# No Disease = 0

y_true = df["Actual"].map({
    "Disease":1,
    "No Disease":0
})

y_pred = df["Predicted"].map({
    "Disease":1,
    "No Disease":0
})
```

```python
#Confusion Matrix

cm = confusion_matrix(y_true, y_pred)

print(cm)   

#             Predicted
#            0      1

# Actual 0   9      3
# Actual 1   2      6     TN=9, FP=3, FN=2, TP=6 
```

```python
#Accuracy
accuracy = accuracy_score(y_true, y_pred)

print("Accuracy =", accuracy)
```

```python
#Precision

precision = precision_score(y_true, y_pred)

print("Precision =", precision)
```

```python
#Recall

recall = recall_score(y_true, y_pred)

print("Recall =", recall)
```

```python
#F1 Score
f1 = f1_score(y_true, y_pred)

print("F1 Score =", f1)
```

```python
#ROC-AUC Score
roc = roc_auc_score(y_true, y_pred)

print("ROC AUC =", roc)
```
