```python
#Stop Word Removal

# is
# am
# are
# the
# a
# an
# of
# to
# in
# for
# and


from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = "I am learning Natural Language Processing"

stop_words = set(stopwords.words("english"))

tokens = word_tokenize(text)

filtered = [word for word in tokens if word.lower() not in stop_words]

print(filtered)
```

#Stemming

Stemming reduces words to their root form by removing prefixes or suffixes using simple rules. The resulting word (called a stem) may not be a valid dictionary word.

```python
# | Original Word | Stem    |
# | ------------- | ------- |
# | Playing       | Play    |
# | Played        | Play    |
# | Plays         | Play    |
# | Studies       | Studi   |
# | Running       | Run     |
# | Connected     | Connect |
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["mornings", "played", "studies", "running"]

for word in words:
    print(word, "→", stemmer.stem(word))
```

#Lemmatization

Lemmatization converts words into their dictionary (base) form, called the lemma, using vocabulary and grammar.

```python
# | Original Word | Lemma |
# | ------------- | ----- |
# | Running       | Run   |
# | Studies       | Study |
# | Better        | Good  |
# | Mice          | Mouse |
# | Children      | Child |
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["running", "studies", "better", "mice"]

for word in words:
    print(word, "→", lemmatizer.lemmatize(word))
```
