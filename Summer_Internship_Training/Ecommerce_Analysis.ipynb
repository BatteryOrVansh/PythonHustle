# **DATA CLEANING** ✅

```python
print("Check Missing Values:\n", df.isnull())
```

```python
print("Total missing values:\n", df.isnull().sum())
```

```python
df = df.dropna(subset=["Description"])
df.isnull().sum()
```

```python
df_with_customer = df.dropna(subset=["CustomerID"]).copy()
df_with_customer["CustomerID"] = df_with_customer["CustomerID"].astype(int)

df_with_customer
```

```python
df.isnull().sum()
```

```python
df_with_customer.isnull().sum()
```

```python
df["IsCancelled"] = df["InvoiceNo"].astype(str).str.startswith("C")
df["IsCancelled"]
```

```python
# Create a new dataframe only keepig valid transaction
df_clean = df[
    (~df["IsCancelled"]) & # keep only rows where IsCancelled is false, means confirm orders
    (df["Quantity"] > 0) & # keep only those rows where quamtity is greater than 0
    (df["UnitPrice"] > 0) # keep those rows where price is greater than 0
].copy() # create a seperate copy of the filtered dataframe
```

```python
df_clean
```

```python
# pd.to_datetime is used to convert into standardized pandas datetime objects

df_clean["InvoiceDate"] = pd.to_datetime(df_clean["InvoiceDate"])
df["InvoiceDate"]
```

```python
# Create a new column "Total Price" to calculate the total amnt for each txn
# Total Price = qty * unitPrice
# we use np.multiply() to perform element wise multiplication between the two col

df_clean["TotalPrice"] = np.multiply(df_clean["Quantity"], df_clean["UnitPrice"])
df_clean["TotalPrice"]
```

```python
print("STEP3: Data Cleaned")
print("Original Rows: ", df.shape[0], "| Clean rows: ", df_clean.shape[0])
print("Cancelled orders removed: ", df["IsCancelled"].sum())
print(df_clean[["InvoiceNo", "Quantity", "UnitPrice", "TotalPrice", "InvoiceDate"]])
print("\n")
```

# STEP 4: Date-based feature (for time analysis)

```python
# Extract the year from the "InvoiceDate" column and store it in a new column "Year"
df_clean["Year"] = df_clean["InvoiceDate"].dt.year

#Extract the month (1-12 from the "InvoiceDate" col
df_clean["Month"] = df_clean["InvoiceDate"].dt.month

# Extract the day of the month (1-31) from the "InvoiceDate" column
df_clean["Day"] = df_clean["InvoiceDate"].dt.day

# Extract the day of the week (0-6) from the "InvoiceDate" column
df_clean["DayOfWeek"] = df_clean["InvoiceDate"].dt.dayofweek

#Extract the hour (0-23) from the "InvoiceDate"
df_clean["Hour"] = df_clean["InvoiceDate"].dt.hour

#Extract the name of the day of the week
df_clean["DayOfWeekName"] = df_clean["InvoiceDate"].dt.day_name()
```

# STEP 5: **KEY BUSINESS METRICS**

```python
# Calculate the total rvenue generated frokm all valid transacgtions.
# np.sum() adds all values in the 'TotalPrice' column
# Forumla: Total Revenue: Sum of all totalPrice values

total_revenue = np.sum(df_clean["TotalPrice"])

# Calculate the total number of unique orders (invoices)
# nunique() counts unique InvoiceNo values
# If an invoice contains multiple producrs, it is still counted as ONE order
total_orders = df_clean["InvoiceNo"].nunique()


# Calculate the total number of unique customers.
# nunique() counts distinct CustomerID values.
# If a customer places multiple orders, they are counted only One
total_customers = df_clean["CustomerID"].nunique()

# Calculate the average order value (AOV)
# step1: group all rows by InvoiceNO (each invoice may contain multiple products)
# step2: sum the TotalPrice of all products within each invoice
# step3: calculate the average of all invoice totals
avg_order_value = df_clean.groupby("InvoiceNo")["TotalPrice"].sum().mean()
```

```python
print("Step 5: Key Metrics")
print("Total Revenue: ", total_revenue)
print("Total Orders: ", total_orders)
print("Total Customers: ", total_customers)
print("Average Order Value: ", avg_order_value)
```

# **STEP 6: TOP PRODUCTS (by revenue and by quantity)**

```python
# Find the top 10 products that generated the highest total revenue
top_products_revenue = (
    df_clean.groupby("Description")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# find the top 10 products sold in the highest quantities
top_products_quantity = (
    df_clean.groupby("Description")["Quantity"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
# Display the reuslt
print("Top 10 Products by Revenue:\n", top_products_revenue)
print("\nTop 10 Products by Quantity Sold:\n", top_products_quantity)
```

# **STEP 7: COUNTRY-WISE ANALYSIS**

```python
# Calculate the total revenue generated by each country and sort it from hightest to lowest
country_revenue = (
    df_clean.groupby("Country")["TotalPrice"]
    .sum()
    .astype(int)
    .sort_values(ascending=False)
)
```

```python
# Display the result
print("Step 7: Revenue by Country (Top10)")
print(country_revenue.head(10))
print("\n")
```

# **STEP 8: MONTHLY SALES TREND**

```python
# Create a new column containing only the year and month from the InvoiceDate
df_clean["YearMonth"] = df_clean["InvoiceDate"].dt.to_period("M")

# Calculate the total sales (revenue) for each Year-Month
monthly_sales = df_clean.groupby("YearMonth")["TotalPrice"].sum()
```

```python
# Display the result

print("Step 8: Monthly revenue trend")
print(monthly_sales)
print("\n")
```
